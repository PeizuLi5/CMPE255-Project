# Data Spliting, Training, Testing, and Evaluation

```python
# Importing scikit-learn libraries
import sklearn
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso

from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# Splitting the DataFrame into training and testing sets
# train_size=0.9 indicates that 90% of the data will be used for training, and the remaining 10% for testing.
# random_state is set for reproducibility of the split.
df_train, df_test = train_test_split(df, train_size=0.9, random_state = 100)
```

```python
print("X train data size ->", df_train.shape)
print("X test data size ->", df_test.shape)
```

```python
# Removing the 'price' column from the list
scale_columns.remove('price')
```

```python
# Creating a StandardScaler instance
scaler = StandardScaler()
# Fitting and transforming the selected columns in the training set
df_train[scale_columns] = scaler.fit_transform(df_train[scale_columns])
# Transforming the selected columns in the testing set using the scaler fitted on the training set
df_test[scale_columns] = scaler.transform(df_test[scale_columns])
```

```python
# Creating the target variable (y_train) by popping the 'price' column from the training set
y_train = df_train.pop('price')
# Creating the feature matrix (X_train) by excluding the 'price' column from the training set
X_train = df_train
```

```python
# Creating the target variable (y_test) by popping the 'price' column from the testing set
y_test = df_test.pop('price')
# Creating the feature matrix (X_test) by excluding the 'price' column from the testing set
X_test = df_test
```

```python
def model_eval(model):
    # Lets calculate some metrics such as R2 score, RSS and RMSE
    # Predictions on the training and testing sets
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    # List to store evaluation metrics
    metric = []
    
    # R2 scores
    r2_train_lr = r2_score(y_train, y_pred_train)
    print("R2 score for train:", r2_train_lr)
    metric.append(r2_train_lr)

    r2_test_lr = r2_score(y_test, y_pred_test)
    print("R2 score for test:", r2_test_lr)
    metric.append(r2_test_lr)
    print("--"*20)
    
    # Residual Sum of Squares (RSS)
    rss1_lr = np.sum(np.square(y_train - y_pred_train))
    print("RSS for train:", rss1_lr)
    metric.append(rss1_lr)

    rss2_lr = np.sum(np.square(y_test - y_pred_test))
    print("RSS for test:", rss2_lr)
    metric.append(rss2_lr)
    print("--"*20)

    mse_train_lr = mean_squared_error(y_train, y_pred_train)
    print("MSE for train:", mse_train_lr)
    metric.append(mse_train_lr)

    mse_test_lr = mean_squared_error(y_test, y_pred_test)
    print("MSE for test:", mse_test_lr)
    metric.append(mse_test_lr)
    print("--"*20)
    
    rmse_train_lr = mse_train_lr**0.5
    print("RMSE for train:", rmse_train_lr)
    metric.append(rmse_train_lr)
    
    rmse_test_lr = mse_test_lr**0.5
    print("RMSE for test:", rmse_test_lr)
    metric.append(rmse_test_lr)
    print("--"*20)
    
    return metric, y_pred_train, y_pred_test
```

```python
# Create linear regression object
reg = LinearRegression()
reg.fit(X_train, y_train)
```
